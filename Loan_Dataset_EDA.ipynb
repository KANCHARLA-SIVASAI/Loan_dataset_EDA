```python
object_data=df.select_dtypes(include='object').columns#to get only object type data
for col in object_data:
  df[col].fillna(df[col].mode()[0],inplace=True)#filling null values of categorical data with mode
df
```

```python
df.isna().sum()#to get count of null values in columnwise
```

```python
num_features=df.select_dtypes(include='number').columns#selecting only numeric data
num_features
```

```python
fig,axes=plt.subplots(3,3,figsize=(15,15))
axes=axes.flatten()
for i,col in enumerate(num_features):
  sns.boxplot(x=df[col],ax=axes[i])
  axes[i].set_title(f'boxplot for {col}')
plt.tight_layout()
```

```python
for col in num_features:
  df[col].fillna(df[col].median(),inplace=True)#filling null values in numerical data with median because they have outliers.
df
```

```python
df.isna().sum()
```

```python
num_features1=num_features[:-1]#selecting all numerical cols except loan_status because it have 0 or 1 options only
num_features1
```

```python
#handling outliers
for col in num_features1:
  q1=df[col].quantile(0.25)
  q3=df[col].quantile(0.75)
  iqr=q3-q1
  lower_limit=q1-1.5*iqr
  upper_limit=q3+1.5*iqr
  df[col]=np.where(df[col]<lower_limit,lower_limit,df[col])
  df[col]=np.where(df[col]>upper_limit,upper_limit,df[col])
df
```

```python
#to check outliers handled or not
fig,axes=plt.subplots(4,2,figsize=(10,20))
axes=axes.flatten()
for i,col in enumerate(num_features1):
  sns.boxplot(x=df[col],ax=axes[i])
  axes[i].set_title(f'boxplot for {col}')
plt.tight_layout()
```

#conclusion:
The exploratory data analysis (EDA) of the loan dataset has provided several valuable insights:

Data Quality and Cleaning
The dataset contained some missing values and outliers, particularly in numeric columns like loan_amount and annual_income. These were successfully handled using imputation and IQR-based filtering, ensuring a cleaner dataset for analysis.

Demographic Insights

The gender distribution shows a moderate imbalance, with one group slightly more represented.

Age distribution suggests most applicants fall within the working-age population, with some skewness toward younger or middle-aged individuals.

Loan and Financial Trends

The majority of loan applicants fall within a mid-to-low income bracket.

Higher loan amounts correlate moderately with higher incomes, which is expected.

Certain types of loans or purposes are more frequently applied for, as shown by categorical plots.

Target Variable Analysis (loan_status)

Class imbalance may be present (e.g., more approved loans than defaults).

Visuals comparing features with loan_status can help identify key factors affecting approval.

Feature Relationships
The heatmap reveals several notable correlations. While no extreme multicollinearity was found, there are moderate relationships worth considering during feature selection for modeling.
